# K-Means

```python
dataset_questions = dataset.iloc[:,5:33]
```

```python
dataset_questions.head()
```

```python
# PCA
from sklearn.decomposition import PCA
pca = PCA(n_components = 4)
dataset_questions_pca = pca.fit_transform(dataset_questions)
print(pca.explained_variance_ratio_)
```

可以看到用前两个特征可以表达86%左右的特征，因此降维到2维

```python
pca = PCA(n_components = 2)
dataset_questions_pca = pca.fit_transform(dataset_questions)
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 12):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(dataset_questions_pca)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 12), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# 根据Elbow原则使用n=3
distortions=[]
for i in range(1, 15):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    y_kmeans = kmeans.fit_predict(dataset_questions_pca)
    distortions.append(kmeans.inertia_)
print(distortions)
```

```python
# CH系数
from sklearn import metrics
kmeans = KMeans(n_clusters = 3, init = 'k-means++')
y_kmeans = kmeans.fit_predict(dataset_questions_pca)
labels = kmeans.labels_
metrics.calinski_harabasz_score(dataset_questions_pca, labels)
```

```python
# 轮廓系数
from sklearn import metrics
kmeans = KMeans(n_clusters = 3, init = 'k-means++')
y_kmeans = kmeans.fit_predict(dataset_questions_pca)
labels = kmeans.labels_
metrics.silhouette_score(dataset_questions_pca, labels, metric='euclidean')
```

```python
# Visualising the clusters
plt.scatter(dataset_questions_pca[y_kmeans == 0, 0], dataset_questions_pca[y_kmeans == 0, 1], s = 20, c = 'yellow', label = 'Cluster 1')
plt.scatter(dataset_questions_pca[y_kmeans == 1, 0], dataset_questions_pca[y_kmeans == 1, 1], s = 20, c = 'green', label = 'Cluster 2')
plt.scatter(dataset_questions_pca[y_kmeans == 2, 0], dataset_questions_pca[y_kmeans == 2, 1], s = 20, c = 'red', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 20, c = 'blue', label = 'Centroids')
plt.title('Clusters of students')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```

# Hierarchy

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(dataset_questions_pca, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('questions')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
# Fitting Hierarchical Clustering to the dataset
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(dataset_questions_pca)
X = dataset_questions_pca

#print(y_hc)
# 轮廓系数
from sklearn import metrics
# kmeans = KMeans(n_clusters = 3, init = 'k-means++')
# y_kmeans = kmeans.fit_predict(dataset_questions_pca)
# labels = kmeans.labels_
test = metrics.silhouette_score(dataset_questions_pca, y_hc, metric='euclidean')
print(test)

# Visualising the clusters
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 20, c = 'yellow', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 20, c = 'red', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 20, c = 'blue', label = 'Cluster 3')
plt.title('Clusters of STUDENTS')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```
